# Cleaning Data Using pandas

# An Excel File

```python
import pandas as pd  # Importing Pandas Library
```

```python
a = "Customer Call List.xlsx"  # DataSet
```

```python
df=pd.read_excel(a)
```

```python
df
```

# Removing duplicate 

```python
df=df.drop_duplicates()
df
```

# Deleting the Not-Useful-Col

```python
df=df.drop(columns="Not_Useful_Column")
df
```

## Removing the symbols like /, ..., _ etc from name col

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
# or simple you also use that
# df["Last_Name"] = df["Last_Name"].str.rstrip("123._/")
df
```

```python
df
```

### Now cleaning the phone_numbers you see that there is a straight Integer like this 876|678|3469 having brakets nan values so we can clear it to a correct phone Number squence like this 123-545-5421 

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True) # RegEX mean return any value 
                                                                                 # except a-zA-Z0-9 replace with nothing ''.
df["Phone_Number"]

# This modification uses regex=True to ensure that the provided pattern is treated as a regular expression. 
```

```python
df
```

## This code snippet applies a lambda function that converts each element in the "Phone_Number" column to a string. If the values were not already strings, this ensures that they are treated as such.

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x)) # coverting into string
df["Phone_Number"]
```

## If you want to add hyphens to create a specific format (e.g., xxx-xxx-xxxx), you can modify your code as follows:

### python
### df["Phone_Number"] = df["Phone_Number"].apply(lambda x: f'{x[:3]}-{x[3:6]}-{x[6:]}')
### This code uses f-strings to create the desired format by slicing the original string into three parts. It assumes that the phone number has at least 10 characters. If some entries might have fewer characters, you may need to add additional checks to handle those cases.

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"]
```

```python
df
```

### . Now Replacing 'nan--	' by '  '

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--', '')
df["Phone_Number"]
```

```python
df
```

```python
df["Address"]
```

## The str.split() method is used to split a string into a list of substrings based on a specified delimiter.
## This uses the keyword argument n to specify the maximum number of splits. The resulting DataFrame will have three columns ('Street_Address', 'State', 'ZipCode') based on the split results. If there are fewer than three parts in the address, the remaining columns will contain NaN values.

```python
df[['Street_Address', 'State', 'ZipCode']] = df["Address"].str.split(',', n=2, expand=True)
df[['Street_Address', 'State', 'ZipCode']]
```

```python
df
```

```python
df['Paying Customer']
```

### We want to change the value of Yes into Y and No into No

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df['Paying Customer']
```

```python
df
```

### Similarly in Do_Not_Contact also We Change Like obove

```python
df["Do_Not_Contact"]
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact']
```

```python
df
```

## At Last removing the NaN, N/a and none values from our dataset

```python
df=df.replace('N/a', '')
df
```

```python
df=df.fillna('')
df
```

### This code snippet iterates through the index of a DataFrame df and checks the value in the 'Do_Not_Contact' column for each row. If the value is 'Y' (indicating that the contact should not be made), it removes the corresponding row from the DataFrame using the drop method with inplace=True. After the loop completes, the modified DataFrame is accessible as df.

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
        
df
```

### Similar for Phone Number

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
        
df
```

### At last rest the Indexes

```python
df=df.reset_index(drop=True)
df
```
